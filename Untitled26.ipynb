```python
from sklearn.decomposition import PCA

# Select only numerical columns for PCA
numerical_cols = df1.select_dtypes(include=['number']).columns.tolist()

# Exclude the original 'DATE' column if it's still in numerical format after label encoding
if 'DATE' in numerical_cols:
    numerical_cols.remove('DATE')

# Exclude the newly created normalized and standardized columns from the features for PCA
exclude_cols = ['A31ANO_normalized', 'A31ANO_standardized']
numerical_cols = [col for col in numerical_cols if col not in exclude_cols]


X = df1[numerical_cols].copy() # Create a copy to avoid modifying the original df1 unintentionally

# Fill NaNs with column mean (or median)
X = df1[numerical_cols].copy()
X = X.fillna(X.mean())   # or X.fillna(0) or sklearn.impute.SimpleImputer

# Drop rows with NaN values from X before applying PCA
#X.dropna(inplace=True)

# Initialize PCA - you can specify the number of components or leave it as None to get all components
pca = PCA()

# Fit PCA to the data
principal_components = pca.fit_transform(X)

# Create a new DataFrame with the principal components
pca_df = pd.DataFrame(principal_components, columns=[f'principal_component_{i+1}' for i in range(principal_components.shape[1])])

# Concatenate the principal components with the original DataFrame (optional, depending on your needs)
# Need to reset index of both dataframes before concatenating if rows were dropped from X
df1_reset = df1.loc[X.index].reset_index(drop=True)
pca_df_reset = pca_df.reset_index(drop=True)
df1 = pd.concat([df1_reset, pca_df_reset], axis=1)


display(df1.head())

# Display explained variance ratio
print("Explained variance ratio:")
print(pca.explained_variance_ratio_)

# Display cumulative explained variance ratio
print("\nCumulative explained variance ratio:")
print(pca.explained_variance_ratio_.cumsum())
```

```python
# Divide the time series data into three parts
n = len(df1)
part_size = n // 3

part1 = df1.iloc[:part_size]
part2 = df1.iloc[part_size:2*part_size]
part3 = df1.iloc[2*part_size:]

# Calculate mean and variance for each part for the 'A31ANO' column
mean1 = part1['A31ANO'].mean()
mean2 = part2['A31ANO'].mean()
mean3 = part3['A31ANO'].mean()

variance1 = part1['A31ANO'].var()
variance2 = part2['A31ANO'].var()
variance3 = part3['A31ANO'].var()

print(f"Mean of Part 1: {mean1:.2f}")
print(f"Mean of Part 2: {mean2:.2f}")
print(f"Mean of Part 3: {mean3:.2f}")

print(f"\nVariance of Part 1: {variance1:.2f}")
print(f"Variance of Part 2: {variance2:.2f}")
print(f"Variance of Part 3: {variance3:.2f}")

# Check for stationarity (simple comparison of means and variances)
if mean1 != mean2 and mean2 != mean3 and variance1 != variance2 and variance2 != variance3:
    print("\nBased on this simple test, the time series appears to be non-stationary.")
else:
    print("\nBased on this simple test, the time series appears to be relatively stationary or the differences are not significant enough by this method.")
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform the ADF test on the 'A31ANO' column
result = adfuller(df1['A31ANO'])

# Extract and print the results
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))

# Interpret the results
if result[1] <= 0.05:
    print("\nReject the null hypothesis (H0), the time series is stationary.")
else:
    print("\nFail to reject the null hypothesis (H0), the time series is non-stationary.")
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Decomposition
# The model can be 'additive' or 'multiplicative'
# Additive model: Time Series = Trend + Seasonality + Residual
# Multiplicative model: Time Series = Trend * Seasonality * Residual
# Let's use the additive model for now.

# Drop rows with NaN values in the 'A31ANO' column before decomposition
df1_cleaned = df1.dropna(subset=['A31ANO'])

# Perform seasonal decomposition on the cleaned data
decomposition = seasonal_decompose(df1_cleaned['A31ANO'], model='additive', period=12) # period is 12 for monthly data

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plot the decomposed components
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(df1_cleaned['DATE'], df1_cleaned['A31ANO'], label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(df1_cleaned['DATE'], trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(df1_cleaned['DATE'], seasonal, label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(df1_cleaned['DATE'], residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
plt.show()

# To detrend using decomposition, subtract the trend component
# Ensure alignment by using the index from the trend
df1_cleaned['A31ANO_detrended_decomp'] = df1_cleaned['A31ANO'] - trend

plt.figure(figsize=(12, 6))
plt.plot(df1_cleaned['DATE'], df1_cleaned['A31ANO_detrended_decomp'])
plt.xlabel('DATE')
plt.ylabel('Detrended A31ANO Value (Decomposition)')
plt.title('A31ANO Detrended using Decomposition')
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np
import matplotlib.pyplot as plt

# Exponential Smoothing
# Drop rows with NaN values in the 'A31ANO' column before smoothing
df1_cleaned_smooth = df1.dropna(subset=['A31ANO'])

# Fit the Exponential Smoothing model
# Let's use a simple exponential smoothing model (no trend, no seasonality) for initial smoothing
model = ExponentialSmoothing(df1_cleaned_smooth['A31ANO'], trend=None, seasonal=None)
fit = model.fit(smoothing_level=0.2) # You can experiment with different smoothing_level values

# Get the smoothed values
df1_cleaned_smooth['A31ANO_smoothed_exp'] = fit.fittedvalues

# Plot original and smoothed data to visually inspect for linearity
plt.figure(figsize=(12, 6))
plt.plot(df1_cleaned_smooth['DATE'], df1_cleaned_smooth['A31ANO'], label='Original A31ANO', alpha=0.7)
plt.plot(df1_cleaned_smooth['DATE'], df1_cleaned_smooth['A31ANO_smoothed_exp'], label='Exponential Smoothing', color='red')
plt.xlabel('DATE')
plt.ylabel('A31ANO Value')
plt.title('A31ANO: Original vs. Exponential Smoothing')
plt.legend()
plt.grid(True)
plt.show()

# Checking for Linearity
# Method 1: Visual Inspection (from the plot above)
print("\nVisual Inspection for Linearity:")
print("Observe the plot above. If the smoothed line appears to follow a roughly straight path, it might suggest a linear trend.")
print("Deviations from a straight line indicate non-linearity.")

# Method 2: Fit a simple linear regression model and check R-squared
# Use the numerical representation of the date for regression
# We can use the index or the 'date_ordinal' column created earlier
X = df1_cleaned_smooth['date_ordinal'].values.reshape(-1, 1)
y = df1_cleaned_smooth['A31ANO'].values

# Fit the linear regression model
linear_model = LinearRegression()
linear_model.fit(X, y)

# Predict using the linear model
y_pred_linear = linear_model.predict(X)

# Calculate R-squared
r2 = r2_score(y, y_pred_linear)

print(f"\nR-squared from Simple Linear Regression: {r2:.4f}")
print("R-squared value indicates the proportion of the variance in the dependent variable (A31ANO) that is predictable from the independent variable (Time).")
print("A higher R-squared value (closer to 1) suggests a stronger linear relationship.")
print("A lower R-squared value suggests that a simple linear model does not capture the variation well, indicating potential non-linearity.")

# Method 3: Plot the linear regression fit
plt.figure(figsize=(12, 6))
plt.scatter(df1_cleaned_smooth['DATE'], df1_cleaned_smooth['A31ANO'], label='Original A31ANO', alpha=0.5)
plt.plot(df1_cleaned_smooth['DATE'], y_pred_linear, color='green', label='Linear Regression Fit')
plt.xlabel('DATE')
plt.ylabel('A31ANO Value')
plt.title('A31ANO: Original vs. Linear Regression Fit')
plt.legend()
plt.grid(True)
plt.show()

print("\nInterpretation:")
print("- If the linear regression line in the plot fits the data points well, it suggests a linear relationship.")
print("- If the data points deviate significantly from the linear regression line, it indicates non-linearity.")
print("- Combined with the R-squared value and the visual inspection of the smoothed data, you can get a better understanding of the linearity of the time series.")
```
