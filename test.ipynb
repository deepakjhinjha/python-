```python
# Program to find LCM using while loop

# Input numbers
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Start with the greater of the two numbers
lcm = max(num1, num2)

# Loop until we find a number divisible by both
while True:
    if lcm % num1 == 0 and lcm % num2 == 0:
        print("LCM of", num1, "and", num2, "is", lcm)
        break
    lcm += 1
```

```python
# Program to check if a number is Perfect using while loop

num = int(input("Enter a number: "))

# Initialize variables
i = 1
sum_of_divisors = 0

# Loop through all proper divisors
while i < num:
    if num % i == 0:
        sum_of_divisors += i
    i += 1

# Check if sum of divisors equals the number
if sum_of_divisors == num:
    print("Perfect")
else:
    print("Not Perfect")
```

```python
# Program to count prime digits in a number using while loop

num = int(input("Enter a number: "))

count = 0
prime_digits = [2, 3, 5, 7]

while num > 0:
    digit = num % 10   # extract last digit
    if digit in prime_digits:
        count += 1
    num //= 10         # remove last digit

print("Count of prime digits:", count)
```

```python
# Program to print Reverse Number Triangle

rows = int(input("Enter number of rows: "))

i = 1
while i <= rows:
    j = i
    while j >= 1:
        print(j, end=" ")
        j -= 1
    print()   # move to next line
    i += 1
```

```python
# Program to print Star + Number Mix Pattern

rows = int(input("Enter number of rows: "))

i = 1
while i <= rows:
    print("*", end=" ")   # print star first
    j = 1
    while j <= i:
        print(j, end=" ")
        j += 1
    print()   # move to next line
    i += 1
```

```python
# Program to print star pattern

rows = 5   # you can change this value

# First half: descending stars
i = rows
while i >= 1:
    print("*" * i)
    i -= 1

# Second half: ascending stars
i = 2
while i <= rows:
    print("*" * i)
    i += 1
```
