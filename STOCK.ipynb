```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Nifty 50 Historical Data.csv')
# taken this dataset from investing.com
# its the data of past one month of nifty fifty stock
```

```python
# Convert date column to datetime

df['Date'] = pd.to_datetime(df['Date'], errors = 'coerce')

```

```python
numeric_cols = ['Price', 'Open', 'High', 'Low', 'Vol.', 'Change %']

for col in numeric_cols:

    df[col] = df[col].astype(str).str.replace(',', '', regex=False).str.strip()


    if col == 'Vol.':
        df[col] = df[col].str.replace('M', '', regex=False)
    elif col == 'Change %':
        df[col] = df[col].str.replace('%', '', regex=False)

    df[col] = pd.to_numeric(df[col], errors='coerce')
```
