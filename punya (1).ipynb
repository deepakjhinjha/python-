```python
import pandas as pd
```

```python
df = pd.read_csv("/content/covid_toy.csv")
```

```python
df.dtypes
```

```python
df1_catg = df.loc[:, df.dtypes == 'object'].copy()
df1_catg
```

```python
df1_catg['gender'] = df1_catg['gender'].replace({'Male':0 , 'Female':1})
df1_catg['cough'] = df1_catg['cough'].replace({'Mild':0 , 'Strong':1})
df1_catg['has_covid'] = df1_catg['has_covid'].replace({'No':0 , 'Yes':1})

df1_catg
```

```python
df2_numri = df.loc[:,df.dtypes != 'object']
df2_numri
```
