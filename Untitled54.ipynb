```python
#reversing the string
def rev(s):
  rev = ""
  for ch in s:
    rev = ch+rev

  return rev
```

```python
print(rev("hello"))
```

```python
# count unique elements
def counts(s):
  return len(set(s))
```

```python
print(counts([1,2,1,2,2,2,3,4,3,3,2,1,3]))
```

```python
def get_unique_elements(arr):
    freq = {}
    for x in arr:
        freq[x] = freq.get(x, 0) + 1

    unique_list = [k for k, v in freq.items() if v == 1]
    return unique_list, len(unique_list)

print(get_unique_elements([1,2,2,3,4,4,5]))
```
